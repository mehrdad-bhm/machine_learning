```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Data Science/4. Clustering (Unsupervised Learning)/Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

### Using Dendogram to find the optimal number of clusters


```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))                          # ward : minimum variance
plt.title('dendrogram')
plt.xlabel('customers')
plt.ylabel('euclidean distances')
plt.show()
```

### Training the Hierarchical model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y = hc.fit_predict(X)
print(y)
```

### Visualising the clusters

```python
plt.scatter(X[y==0, 0], X[y==0, 1], label='Cluster 1') # y==0   -> specify data-points which blong to cluster 0
plt.scatter(X[y==1, 0], X[y==1, 1], label='Cluster 2')
plt.scatter(X[y==2, 0], X[y==2, 1], label='Cluster 3')
plt.scatter(X[y==3, 0], X[y==3, 1], label='Cluster 4')
plt.scatter(X[y==4, 0], X[y==4, 1], label='Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual (k$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
