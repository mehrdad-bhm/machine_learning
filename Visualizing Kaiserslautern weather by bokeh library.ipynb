## 2. Customizing the temperature chart

As detailed above, your visualization should look like a modern version of the one from the New York Times. This can be achieved by changing the graphical elements and styling visual properties.
```python
def create_temperature_chart(df, width=900):
    '''Create a bokeh figure for temperature range and normal values.'''

    # create figure and data source
    p = figure(width=width, height=600, title='Kaiserslautern\'s Weather for 2018\n', tools=['xwheel_zoom'],
           x_axis_type="datetime", x_axis_location="above", y_range=(-15,40))

    source = ColumnDataSource(df)

    # add graphical items
    p.line(source=source, x='Timestamp', y='temp_max', line_color='crimson', legend_label="Max tempe")
    p.line(source=source, x='Timestamp', y='temp_min', line_color='darkblue', legend_label="min tempe")
    p.varea(x='Timestamp', y1='temp_min', y2='temp_max', source=source, fill_color='lightcyan', fill_alpha=0.5)


    # mark min/max temperature
    p.circle(x=df['temp_max'].idxmax() , y= df_kl.at[df['temp_max'].idxmax(),'temp_max'] , source=source, size=10, color='red', legend_label='Max peak temp')
    p.circle(x=df['temp_min'].idxmin() , y= df_kl.at[df['temp_min'].idxmin(),'temp_min'] , source=source, size=10, color='blue', legend_label='min temp')

    #temp normal min/max
    p.line(source=source, x='Timestamp', y='temp_normal_min', line_color='red', legend_label="Normal min tempe", line_dash='dotdash')
    p.line(source=source, x='Timestamp', y='temp_normal_max', line_color='blue', legend_label="Normal Max tempe", line_dash='dashdot')


    # style visual attributes
    p.xaxis.ticker = MonthsTicker(months=list(range(12)))
    p.xgrid.ticker = MonthsTicker(months=list(range(12)))
    p.xaxis.formatter=DatetimeTickFormatter(months="                     %b ")
    p.xaxis.major_label_text_align = 'right'
    p.yaxis[0].formatter = PrintfTickFormatter(format="%2i¬∞")
    p.yaxis.axis_label = "Temperature [¬∞C]"
    p.yaxis.axis_label_text_font_size= "12pt"
    p.xaxis.axis_label = "Mounths"
    p.xaxis.axis_label_text_font_size= "12pt"
    p.title.text_font_size = "15pt"
    p.title.align = "center"
    p.legend.title = 'Lines'

    return p

p = create_temperature_chart(df_kl)
show(p)
```

## 4. Designing additional charts

Now design the charts for precipitation and relative humidity.

Hints for temporal x-axis:
- **Width of bars**: The width is given milliseconds. In order to get the required scaling, you will need to specify the width like: `widthInDays = ndays*24*60*60*1000` (24 hours * 60 minutes * 60 seconds * 1000 milliseconds)
- **Position of bars**: You can shift the bars using the dodge function `x=dodge('prec', value, range=p.x_range)`. Keep in mind that you need to define an appropriate `value` by which to shift the bar.

```python
from bokeh.core.property.color import Alpha
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge
from bokeh.models import ColumnDataSource, Grid, LinearAxis, Plot, VBar


def create_precipitation_chart(df, width=900):

    df['Timestamp'] = pd.to_datetime(df['Timestamp'])

    # Calculate the width of bars in milliseconds
    widthInDays = 12 * 24 * 60 * 60 * 1000

    # Create a ColumnDataSource for the data
    source = ColumnDataSource(df)

    # Create the figure
    p = figure(width=width, height=200, tools=['xwheel_zoom'], x_axis_type="datetime", y_range=(0,150))

    # Plot the bars for precipitation

    #p.line(source=df, x=dodge('Timestamp', -widthInDays/4, range=p.x_range), y='prec')
    p.vbar(x=dodge('Timestamp', widthInDays*0.5, range=p.x_range), top="prec_normal" , source=source, legend_label='prec_normal', color='red',width= widthInDays*0.5)
    p.vbar(x=dodge('Timestamp', 0, range=p.x_range), top="prec", source=source, legend_label='prec', color='blue', width=widthInDays*0.5)
    p.xaxis.ticker = MonthsTicker(months=list(range(12)))
    p.xaxis.formatter=DatetimeTickFormatter(months="%b %Y")
    p.yaxis.axis_label = "Precipitation"
    return p

show(create_precipitation_chart(df_kl_prec))
```

```python
def create_humidity_chart(df, width=900):
 # df['Timestamp'] = pd.to_datetime(df['Timestamp'])

  source = ColumnDataSource(df_kl)
  p = figure(width=width, height=200, tools=['xwheel_zoom'], x_axis_type="datetime")
  p.line(source=source, x='Timestamp', y='rel_humidity')
  p.yaxis.axis_label = "Humidity"
  return p

show(create_humidity_chart(df_kl))
```
